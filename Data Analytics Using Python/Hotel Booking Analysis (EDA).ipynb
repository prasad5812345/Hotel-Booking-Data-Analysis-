```python
#To check the Average Daily Rate(price) of Canceled reservations are high or non-Canceled reservations
canceled_data_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace=True)
canceled_data_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_canceled_data_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace=True)
not_canceled_data_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label='not canceled')
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label='canceled')
plt.legend(fontsize=15)
plt.show()
```

```python
#We will filter the data from 2016 to sept 2017 for both canceled & not_cancele reservations (since the above fig is messy)
canceled_data_adr=canceled_data_adr[(canceled_data_adr['reservation_status_date']>'2016')&(canceled_data_adr['reservation_status_date']<'2017-09')]
not_canceled_data_adr=not_canceled_data_adr[(not_canceled_data_adr['reservation_status_date']>'2016')&(canceled_data_adr['reservation_status_date']<'2017-09')]

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label='not canceled')
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label='canceled')
plt.legend(fontsize=15)
plt.show()
```

```python
bookings_by_months_df =df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns = {'hotel':'Counts'})

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

bookings_by_months_df['arrival_date_month'] = pd.Categorical(bookings_by_months_df['arrival_date_month'], categories = months, ordered = True)

bookings_by_months_df = bookings_by_months_df.sort_values('arrival_date_month')

bookings_by_months_df
```

```python
# Set plot size
plt.figure(figsize = (14,6))

# Plotting lineplot on x- months & y- bookings counts
sns.lineplot(x = bookings_by_months_df['arrival_date_month'], y = bookings_by_months_df['Counts'])

# Set title
plt.title('Number of bookings across each month', fontsize = 20)

# Set labels
plt.xlabel('Month', fontsize = 16)
plt.ylabel('Number of bookings', fontsize = 16)

# To show
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df = df.iloc[0:2999]
df.head()
```

```python
df_new = df.copy()[['required_car_parking_spaces','lead_time','booking_changes','adr','adults', 'is_canceled']]
df_new.head()
```

```python
x = df_new.drop(['is_canceled'], axis=1)
y = df_new['is_canceled']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, shuffle=False)
```

```python
model_LogReg=LogisticRegression()
model_LogReg.fit(x_train,y_train)
print(model_LogReg.coef_)
print(model_LogReg.intercept_)

m=model_LogReg.coef_[0][0]
c=model_LogReg.intercept_[0]
```

```python
y_pred_lr=model_LogReg.predict(x_test)
```

```python
coba=x_test.iloc[:1]
model_LogReg.predict(coba)
coba
```

```python
model_LogReg.score(x_train,y_train)
```

```python
model_LogReg.score(x_test,y_test)
```
